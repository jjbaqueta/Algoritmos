7) O peso ideal de uma pessoa está relacionado com a altura e sexo. Sabendo disso, implemente um programa que irá receber como entrada a altura e o sexo de uma certa pessoa. Em seguida, calcule o peso ideal dessa pessoa. Seu programa deve exibir o peso ideal do usuário levando em considerando o sexo informado pelo mesmo. Para realizar essa tarefa utilize as seguintes equações:

    - Peso ideal para homens = (72.7 * altura) - 58
    - Peso ideal para mulheres = (62.1 * altura) - 44.7

```python
%%file lista2Exer7.c

#include <stdio.h>

int main()
{
    char sexo;
    float altura, peso;

    printf("informe seu sexo, (M) para masculino, (F) para feminino\n");
    scanf("%c",&sexo);

    printf("irforme sua altura\n");
    scanf("%f",&altura);

    switch (sexo){
        case 'M':
            peso = (72.7 * altura) - 58;
            printf("peso ideal: %f\n", peso);
        break;
            
        case 'F':
            peso = (62.1 * altura) - 44.7;
            printf("peso ideal: %f\n", peso);
        break;

        default:
            printf("sexo informado não é válido\n");    
    }

    return 0;
}
```
