1) Determine os resultados das expressões lógicas abaixo, sabendo que as variáveis a, b, e c contêm, respectivamente, os valores 2, 7 e 3. Utilize as constantes 1 e 0 para representar os valores lógicos verdadeiro (V) e falso (F), respectivamente.
    - a < b || c > 0 && 1
    - !!!(c == c) || !0 || b > a
    - a != b && 1 && !0 || c > a
    - !(1 && a > 0) || !(c > a)
    - !(b > a && a <= c && b >= c)

```python
%%file lista6Exer1.c

#include <stdio.h>

int main()
{
    int a = 2, b = 7, c = 3;

    printf("%d\n", a < b || c > 0 && 1);            // 1 || 1 && 1       ->  1 || 1      -> 1
    printf("%d\n", !!!(c == c) || !0 || b > a);     // !!!(1) || 1 || 1  ->  0 || 1 || 1 -> 1
    printf("%d\n", a != b && 1 && !0 || c > a);     // 1 && 1 && 1 || 1  ->  1 || 1      -> 1
    printf("%d\n", !(1 && a > 0) || !(c > a));      // !(1 && 1) || !(1) ->  0 || 0      -> 0
    printf("%d\n", !(b > a && a <= c && b >= c));   // !(1 && 1 && 1)    ->    !1        -> 0

    return 0;
}
```

7) O peso ideal de uma pessoa está relacionado com a altura e sexo. Sabendo disso, implemente um programa que irá receber como entrada a altura e o sexo de uma certa pessoa. Em seguida, calcule o peso ideal dessa pessoa. Seu programa deve exibir o peso ideal do usuário levando em considerando o sexo informado pelo mesmo. Para realizar essa tarefa utilize as seguintes equações:

    - Peso ideal para homens = (72.7 * altura) - 58
    - Peso ideal para mulheres = (62.1 * altura) - 44.7

```python
%%file lista6Exer7.c

#include <stdio.h>

int main()
{
    char sexo;
    float altura, peso;

    printf("informe seu sexo, (M) para masculino, (F) para feminino\n");
    scanf("%c",&sexo);

    printf("irforme sua altura\n");
    scanf("%f",&altura);

    switch (sexo){
        case 'M':
            peso = (72.7 * altura) - 58;
            printf("peso ideal: %f\n", peso);
        break;
            
        case 'F':
            peso = (62.1 * altura) - 44.7;
            printf("peso ideal: %f\n", peso);
        break;

        default:
            printf("sexo informado não é válido\n");    
    }

    return 0;
}
```

Seu programa irá receber como entrada os três lados de um triângulo e exibir se o triângulo é escaleno, isósceles ou equilátero.

```python
%%file lista6Exer12.c

#include <stdio.h>

int main(){

    float lado1, lado2, lado3;
    
    printf("informe o lado 1 (AB)\n");
    scanf("%f",&lado1);

    printf("informe o lado 2 (AC)\n");
    scanf("%f",&lado2);

    printf("informe o lado 3 (BC)\n");
    scanf("%f",&lado3);

    if (lado1 == lado2 && lado2 != lado3 || lado1 == lado3 && lado3 != lado2 || lado2 == lado3 && lado3 != lado1){
        printf("isósceles\n");
    }
    else if (lado1 != lado2 && lado2 != lado3){
        printf("escaleno\n");
    }
    else { // último caso: lado1 == lado2 && lado2 == lado3
        printf("equilátero\n");
    }
       
    return 0;
}
```

14) Um tabuleiro de xadrez possui no total 64 casas. Cada lado do tabuleiro possui 8 casas. Uma das mais famosas peças do xadrez é o cavalo, o qual pode pular sobre outras peças executando sempre um movimento no formato da letra L, tal como exibido abaixo. Sabendo disso, escreva um programa para exibir as possíveis posições que o cavalo pode tomar considerando uma dada posição inicial (x, y). Seu programa também deve exibir o número de jogadas possíveis (possíveis movimentos).

```python
%%file lista6Exer14.c

#include <stdio.h>

int main(){

    int x, y, px, py, contador = 0;
    
    printf("informe a posição inicial do cavalo: coordenadas (x = [1 .. 8], y = [1 .. 8])\n");
    scanf("%d %d",&x, &y);

    if (x < 1 || x > 8 || y < 1 || y > 8){
        printf("a posição inicial do cavalo não é válida\n");
    }
    else{
        
        px = x - 1, py = y + 2;

        if (px >= 1 && py <= 8){
            printf("posição possível: (x = %d, y = %d)\n", px, py);
            contador = contador + 1;
        }

        px = x - 1, py = y - 2;

        if (px >= 1 && py >= 1){
            printf("posição possível: (x = %d, y = %d)\n", px, py);
            contador = contador + 1;
        }

        px = x - 2, py = y + 1;

        if (px >= 1 && py <= 8){
            printf("posição possível: (x = %d, y = %d)\n", px, py);
            contador = contador + 1;
        }

        px = x - 2, py = y - 1;

        if (px >= 1 && py >= 1){
            printf("posição possível: (x = %d, y = %d)\n", px, py);
            contador = contador + 1;
        }

        px = x + 1, py = y + 2;

        if (px <= 8 && py <= 8){
            printf("posição possível: (x = %d, y = %d)\n", px, py);
            contador = contador + 1;
        }

        px = x + 1, py = y - 2;

        if (px <= 8 && py >= 1){
            printf("posição possível: (x = %d, y = %d)\n", px, py);
            contador = contador + 1;
        }

        px = x + 2, py = y + 1;

        if (px <= 8 && py <= 8){
            printf("posição possível: (x = %d, y = %d)\n", px, py);
            contador = contador + 1;
        }

        px = x + 2, py = y - 1;

        if (px <= 8 && py >= 1){
            printf("posição possível: (x = %d, y = %d)\n", px, py);
            contador = contador + 1;
        }
        
        printf("número de possibilidades: %d\n", contador);
    }
    
    return 0;
}
```
