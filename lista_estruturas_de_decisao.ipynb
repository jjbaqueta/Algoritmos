13) Escreva um programa onde será informado uma data qualquer como entrada (dia/mês/ano). Considere que cada campo da data será informado separadamente. A partir da data informada seu programa deve calcular e exibir a data do próximo dia. Lembre-se que em anos bissextos o mês de fevereiro tem 29 dias. Um ano será bissexto se for divisível por 400 ou for divisível por 4 e não divisível por 100.

```python
%%file lista2Exer13.c

#include <stdio.h>

int main(){

    int ano, mes, dia;

    printf("informe o ano\n");
    scanf("%d",&ano);

    printf("informe o mês\n");
    scanf("%d",&mes);

    printf("informe o dia\n");
    scanf("%d",&dia);


    if(mes == 1 || mes == 3 || mes == 5 || mes == 7 || mes == 8 || mes == 10){
        if (dia >= 1 && dia < 31){
            dia = dia + 1;
        } 
        else {
            dia = 1;
            mes = mes + 1;
        }
        printf("próximo data: %d/%d/%d\n", dia, mes, ano);
    }
    else if (mes == 4 || mes == 6 || mes == 9 || mes == 11){
        if (dia >= 1 && dia < 30){
            dia = dia + 1;
        } 
        else {
            dia = 1;
            mes = mes + 1;
        }
        printf("próximo data: %d/%d/%d\n", dia, mes, ano);
    }
    else if(mes == 2){
        if ((ano % 400 == 0) || (ano % 4 == 0 && ano % 100 != 0)){
            if (dia >= 1 && dia < 29){
                dia = dia + 1;
            } 
            else {
                dia = 1;
                mes = mes + 1;               
            }
        }
        else {
            if (dia >= 1 && dia < 28){
                dia = dia + 1;
            } 
            else {
                dia = 1;
                mes = mes + 1;               
            }          
        }
        printf("próximo data: %d/%d/%d\n", dia, mes, ano);      
    }
    else if (mes == 12){
        if (dia >= 1 && dia < 31){
            dia = dia + 1;
        } 
        else {
            dia = 1;
            mes = 1;
            ano = ano + 1;
        }
        printf("próximo data: %d/%d/%d\n", dia, mes, ano);
    }
    else {
        printf("data informada não é válida\n");
    }

    return 0;
}
```
