2) Crie um programa que leia um vetor de 30 números inteiros e gere um segundo vetor cujas posições pares são o dobro do vetor original e as ímpares o triplo.

```python
%%file lista8Exerc2.c

#include <stdio.h>

#define N 30

int main() {
    int vetA[N], vetB[N], i;

    // Lendo os valores de entrada - uma valor para cada posição do vetor A

    for (i = 0; i < N; i++) {
        printf("informe o valor %d:\n", i + 1);
        scanf("%d", &vetA[i]);
    }

    // Escrevando no vetor B
    
    for (i = 0; i < N; i++) {
        if (i % 2 == 0) {           // Índice par
            vetB[i] = vetA[i] * 2;
        }
        else {                      // Índice ímpar
            vetB[i] = vetA[i] * 3;
        }
    }

    // Exibindo o conteúdo dos vetores A e B
    
    printf("Vetor A:\n");

    for (i = 0; i < N; i++) {
        printf("%d ", vetA[i]);
    }
    printf("\n");

    printf("Vetor B:\n");

    for (i = 0; i < N; i++) {
        printf("%d ", vetB[i]);
    }
    printf("\n");
    
    return 0;
}
```

3) Elabore um algoritmo capaz de armazenar 10 notas (valores reais) informadas pelo usuário. Após a receber as notas seu programa deve exibir as seguintes informações:
-   Quantidade de notas maiores que 70;
-   Quantidade de notas menores que 40;
-   Quantidade de notas entre 40 e 70;
-   Média das notas informadas.

```python
%%file lista8Exer3.c

#include <stdio.h>

#define N 10

int main() {
    float vet[N], media = 0;
    int i, menor40 = 0, maior70 = 0, entre40_70 = 0;

    // Lendo os valores de entrada - uma valor para cada posição do vetor

    for (i = 0; i < N; i++) {
        printf("informe a nota %d:\n", i + 1);
        scanf("%f", &vet[i]);
    }

    // Contando as notas de acordo com as restrições
    
    for (i = 0; i < N; i++) {
        if (vet[i] < 40) {
            menor40++;
        }
        else if (vet[i] > 70) {
            maior70++;
        }
        else {               // Equivalente: if (vet[i] >= 40 && vet[i] <= 70)
            entre40_70++;
        }
        media += vet[i];
    }
    media /= N;

    // Exibindo as informações de saída
    
    printf("\nNotas informadas:\n");

    for (i = 0; i < N; i++) {
        printf("%.2f ", vet[i]);
    }
    printf("\n\n");

    printf("Quantidade de notas menores que 40: %d\n", menor40);
    printf("Quantidade de notas maiores que 70: %d\n", maior70);
    printf("Quantidade de notas entre 40 e 70: %d\n", entre40_70);
    printf("Média das notas: %.2f\n", media);
  
    return 0;
}
```

4) Escreva um programa que irá receber como entrada dois vetores inteiros de 5 posições. Na sequência, efetue as respectivas operações aritméticas indicadas por outro vetor de 5 posições, sendo esse terceiro vetor um vetor de caracteres. O vetor de caracteres deve ser inicializado em tempo de compilação, ou seja, após sua declaração. Os resultados das operações devem ser armazenados em um quarto vetor.
- Exemplo:
- v1 (int) = {5, 10, 45, 1, 1}
- v_op (char) = {'+', '*', '-', '/', '*'}
- v2 (int) = {4, 6, 5, 2, 2}
- vr (float) = {9, 60, 40, 0.5, 2}

```python
%%file lista8Exer4.c

#include <stdio.h>
#include <stdlib.h>

int main() {
    int v1[5], v2[5], i;
    char v_op[5] = {'+', '*', '-', '/', '*'};
    float vr[5];

    // Lendo os valores de entrada

    printf("OPERANDOS DO VETOR V1:\n");

    for (i = 0; i < 5; i++) {
        printf("informe o operando %d do vetor v1:\n", i + 1);
        scanf("%d", &v1[i]);
    }
    
    printf("\nOPERANDOS DO VETOR V2:\n");

    for (i = 0; i < 5; i++) {
        printf("informe o operando %d do vetor v2:\n", i + 1);
        scanf("%d", &v2[i]);
    }

    // Realizando as operações aritméticas
    
    for (i = 0; i < 5; i++) {
        switch (v_op[i]) {
            case '+':
                vr[i] = v1[i] + v2[i];
            break;

            case '-':
                vr[i] = v1[i] - v2[i];
            break;
    
            case '*':
                vr[i] = v1[i] * v2[i];
            break;

            case '/':
                vr[i] = (float) v1[i] / v2[i];
            break;
            
            default:
                printf("\nOperação aritmética inválida: %c.\nO programa será finalizado.\n", v_op[i]);
                exit(1);    // Termina o programa devido ao erro da operação
        }
    }

    // Exibindo o conteúdo dos vetores

    printf("\nRESULTADO:");
    
    printf("\nv1:\t");

    for (i = 0; i < 5; i++) {
        printf("%d ", v1[i]);
    }

    printf("\nv_op:\t");

    for (i = 0; i < 5; i++) {
        printf("%c ", v_op[i]);
    }

    printf("\nv2:\t");

    for (i = 0; i < 5; i++) {
        printf("%d ", v2[i]);
    }

    printf("\nvr:\t");

    for (i = 0; i < 5; i++) {
        printf("%.2f ", vr[i]);
    }
    printf("\n");

    return 0;
}
```

5) Declare um vetor de 20 posições. Cada posição deve receber como valor o seu índice + 1. Em seguida, troque o 1º elemento com o 11ª, o 2ª com o 12ª, ..., 10ª com o 20ª.

```python
%%file lista8Exer5.c

#include <stdio.h>

#define N 20

int main() {
    int vet[N], aux, i, j;

    // Inicializando o vetor

    for (i = 0; i < N; i++) {
        vet[i] = i + 1;
    }
    
    // Exibindo o conteúdo do vetor antes de inverter seu conteúdo

    for (i = 0; i < N; i++) {
        printf("%d ", vet[i]);
    }
    printf("\n");

    /*
     * Invertendo o conteúdo do vertor
     * i começa em 0 e a cada iteração ele é incrementado em uma únidade
     * j começa em N - 1 e a cada iteração ele é decrementado em uma únidade
     * Chegamos no meio do vetor quando o valor de i é igual ou maior do que o valor j.
     * Nesse caso, as posições de 0 a (N - 1)/2 foram trocadas com as posições de ((N - 1) / 2) + 1 a N - 1
    */

    for (i = 0, j = N - 1; i <= j; i++, j--) {
        
        // Lógica para trocar o conteúdo de vet[i] por vet[j]        
        aux = vet[i];
        vet[i] = vet[j];
        vet[j] = aux;
    }

    // Exibindo o conteúdo do vetor após a inversão

    for (i = 0; i < N; i++) {
        printf("%d ", vet[i]);
    }
    printf("\n");

    return 0;
}
```

6) Elabore um programa que leia 30 valores inteiros e obtenha qual o tamanho da maior sequência consecutiva de números em ordem crescente.

```python
%%file lista8Exer6.c

#include <stdio.h>

#define N 30

int main() {

    int vet[N], i, tamSeqAtual = 1, tamMaiorSeq = 1;
    
    // Lendo os valores de entrada

    for (i = 0; i < N; i++) {
        printf("informe o valor %d:\n", i + 1);
        scanf("%d", &vet[i]);
    }
    
    /*
     * Calculando o tamanho da maior sequência crescente
     * Se o valor atual (vet[i]) é maior que o anterior (vet[i - 1]), estamos lendo uma sequência crescente.
     * Se o valor atual é menor do que o anterior, a última sequência crescente terminou, assim reiniciamos o tamanho da sequência atual.
    */

    for (i = 1; i < N; i++) {
        if (vet[i] > vet[i - 1]) {
            tamSeqAtual++;
        } 
        else {
            tamSeqAtual = 1;
        }
        
        if (tamSeqAtual > tamMaiorSeq) {
            tamMaiorSeq = tamSeqAtual;
        }
    }
    
    // Exibindo as informações de saída

    printf("Tamanho da maior sequência crescente: %d\n", tamMaiorSeq);
    
    return 0;
}
```

8) Escreva um programa para ordenar um vetor de forma crescente. Sugestão: implemente o algoritmo Selection sort.

```python
%%file lista8Exer8.c

#include <stdio.h>
#include <stdlib.h>
#include <time.h>

#define N 10

int main() {

    int vet[N], i, j, menor, indice, aux;
    
    // Inicializando o vetor com números aleatórios - números de 0 a 99
    
    srand(time(NULL));

    for (i = 0; i < N; i++) {
        vet[i] = rand() % 100;
    }

    // Exibindo o vertor vet não ordenado
    
    printf("Vetor não ordenado:\n");

    for (i = 0; i < N; i++) {
        printf("%d ", vet[i]);
    }
    printf("\n\n");
    
    /* 
     * Ordenando o vetor - implementação do Selection sort
     * Passo 1: encontre o menor elemento no vetor
     * Passo 2: troque esse elemento com o primeiro elemento do vetor
     * Passo 3: repita o processo para os n - 1 elementos restantes, troque o menor elemento com o segundo elemento do vetor. 
     * Passo 4: repita o processo para os n - 2 elementos restantes, troque o menor elemento com o terceiro elemento do vetor, e assim por diante.
    */

    for (i = 0; i < N; i++) {
        menor = vet[i];
        indice = i;

        for (j = i + 1; j < N; j++) {
            if (vet[j] < menor) {
                menor = vet[j];
                indice = j;
            }
        }
    
        if (indice > i){
            aux = vet[i];
            vet[i] = vet[indice];
            vet[indice] = aux;
        }
    }

    // Exibindo o vertor vet ordenado
    
    printf("Vetor ordenado:\n");

    for (i = 0; i < N; i++) {
        printf("%d ", vet[i]);
    }
    printf("\n");
      
    return 0;
}
```

9) Implemente o algoritmo da busca binária. Assuma que o valor de busca será informado pelo usuário. Utilize o algoritmo Selection sort do último exercício para ordenar o vetor.

```python
%%file lista8Exer9.c

#include <stdio.h>
#include <stdlib.h>
#include <time.h>

#define N 10

int main() {

    int vet[N], i, j, menor, indice, aux, valorBusca, ini, fim, meio;
    
    // Inicializando o vetor com números aleatórios - números de 0 a 99
    
    srand(time(NULL));

    for (i = 0; i < N; i++) {
        vet[i] = rand() % 100;
    }
    
    /* 
     * Ordenando o vetor - implementação do Selection sort
     * Passo 1: encontre o menor elemento no vetor
     * Passo 2: troque esse elemento com o primeiro elemento do vetor
     * Passo 3: repita o processo para os n - 1 elementos restantes, troque o menor elemento com o segundo elemento do vetor. 
     * Passo 4: repita o processo para os n - 2 elementos restantes, troque o menor elemento com o terceiro elemento do vetor, e assim por diante.
    */

    for (i = 0; i < N; i++) {
        menor = vet[i];
        indice = i;

        for (j = i + 1; j < N; j++) {
            if (vet[j] < menor) {
                menor = vet[j];
                indice = j;
            }
        }
    
        if (indice > i){
            aux = vet[i];
            vet[i] = vet[indice];
            vet[indice] = aux;
        }
    }

    // Exibindo o vertor vet ordenado
    
    printf("Vetor ordenado:\n");

    for (i = 0; i < N; i++) {
        printf("%d ", vet[i]);
    }
    printf("\n\n");

    // Lendo o valor de busca

    printf("Informe o valor que será buscado no vetor:\n");
    scanf("%d", &valorBusca);

    // Realizando a busca
    
    indice = -1;    
    ini = 0;
    fim = N - 1;
    
    while (ini <= fim) {
        meio = ini + (fim - ini) / 2;

        if (valorBusca == vet[meio]) {
            indice = meio;
            break;
        }
        else if (valorBusca < vet[meio]){
            fim = meio - 1;
        }
        else {
            ini = meio + 1;
        }
    }
    
    if (indice == -1) {
        printf("\nValor de busca não está no vetor\n");
    }
    else {
        printf("\nValor de busca encontrado no índice %d do vetor.\n", indice);
    }
     
    return 0;
}
```
