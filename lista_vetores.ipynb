2) Crie um programa que leia um vetor de 30 números inteiros e gere um segundo vetor cujas posições pares são o dobro do vetor original e as ímpares o triplo.

```python
%%file lista8Exerc2.c

#include <stdio.h>

#define N 30

int main() {
    int vetA[N], vetB[N], i;

    // Lendo os valores de entrada - uma valor para cada posição do vetor A

    for (i = 0; i < N; i++) {
        printf("informe o valor %d:\n", i + 1);
        scanf("%d", &vetA[i]);
    }

    // Escrevando no vetor B
    
    for (i = 0; i < N; i++) {
        if (i % 2 == 0) {           // Índice par
            vetB[i] = vetA[i] * 2;
        }
        else {                      // Índice ímpar
            vetB[i] = vetA[i] * 3;
        }
    }

    // Exibindo o conteúdo dos vetores A e B
    
    printf("Vetor A:\n");

    for (i = 0; i < N; i++) {
        printf("%d ", vetA[i]);
    }
    printf("\n");

    printf("Vetor B:\n");

    for (i = 0; i < N; i++) {
        printf("%d ", vetB[i]);
    }
    printf("\n");
    
    return 0;
}
```

3) Elabore um algoritmo capaz de armazenar 10 notas (valores reais) informadas pelo usuário. Após a receber as notas seu programa deve exibir as seguintes informações:
-   Quantidade de notas maiores que 70;
-   Quantidade de notas menores que 40;
-   Quantidade de notas entre 40 e 70;
-   Média das notas informadas.

```python
%%file lista8Exer3.c

#include <stdio.h>

#define N 10

int main() {
    float vet[N], media = 0;
    int i, menor40 = 0, maior70 = 0, entre40_70 = 0;

    // Lendo os valores de entrada - uma valor para cada posição do vetor

    for (i = 0; i < N; i++) {
        printf("informe a nota %d:\n", i + 1);
        scanf("%f", &vet[i]);
    }

    // Contando as notas de acordo com as restrições
    
    for (i = 0; i < N; i++) {
        if (vet[i] < 40) {
            menor40++;
        }
        else if (vet[i] > 70) {
            maior70++;
        }
        else {               // Equivalente: if (vet[i] >= 40 && vet[i] <= 70)
            entre40_70++;
        }
        media += vet[i];
    }
    media /= N;

    // Exibindo as informações de saída
    
    printf("\nNotas informadas:\n");

    for (i = 0; i < N; i++) {
        printf("%.2f ", vet[i]);
    }
    printf("\n\n");

    printf("Quantidade de notas menores que 40: %d\n", menor40);
    printf("Quantidade de notas maiores que 70: %d\n", maior70);
    printf("Quantidade de notas entre 40 e 70: %d\n", entre40_70);
    printf("Média das notas: %.2f\n", media);
  
    return 0;
}
```

4) Escreva um programa que irá receber como entrada dois vetores inteiros de 5 posições. Na sequência, efetue as respectivas operações aritméticas indicadas por outro vetor de 5 posições, sendo esse terceiro vetor um vetor de caracteres. O vetor de caracteres deve ser inicializado em tempo de compilação, ou seja, após sua declaração. Os resultados das operações devem ser armazenados em um quarto vetor.
- Exemplo:
- v1 (int) = {5, 10, 45, 1, 1}
- v_op (char) = {'+', '*', '-', '/', '*'}
- v2 (int) = {4, 6, 5, 2, 2}
- vr (float) = {9, 60, 40, 0.5, 2}

```python
%%file lista8Exer4.c

#include <stdio.h>

int main() {
    int v1[5], v2[5], i;
    char v_op[5] = {'+', '*', '-', '/', '*'};
    float vr[5];

    // Lendo os valores de entrada

    printf("OPERANDOS DO VETOR V1:\n");

    for (i = 0; i < 5; i++) {
        printf("informe o operando %d do vetor v1:\n", i + 1);
        scanf("%d", &v1[i]);
    }
    
    printf("\nOPERANDOS DO VETOR V2:\n");

    for (i = 0; i < 5; i++) {
        printf("informe o operando %d do vetor v2:\n", i + 1);
        scanf("%d", &v2[i]);
    }

    // Realizando as operações aritméticas
    
    for (i = 0; i < 5; i++) {
        switch (v_op[i]) {
            case '+':
                vr[i] = v1[i] + v2[i];
            break;

            case '-':
                vr[i] = v1[i] - v2[i];
            break;
    
            case '*':
                vr[i] = v1[i] * v2[i];
            break;

            case '/':
                vr[i] = (float) v1[i] / v2[i];
            break;
            
            default:
                printf("Operação aritmética inválida!");
        }
    }

    // Exibindo o conteúdo dos vetores

    printf("\nRESULTADO:");
    
    printf("\nv1:\t");

    for (i = 0; i < 5; i++) {
        printf("%d ", v1[i]);
    }

    printf("\nv_op:\t");

    for (i = 0; i < 5; i++) {
        printf("%c ", v_op[i]);
    }

    printf("\nv2:\t");

    for (i = 0; i < 5; i++) {
        printf("%d ", v2[i]);
    }

    printf("\nvr:\t");

    for (i = 0; i < 5; i++) {
        printf("%.2f ", vr[i]);
    }
    printf("\n");

    return 0;
}
```
