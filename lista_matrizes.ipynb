1) Declare uma matriz de inteiros com 10 linhas e 10 colunas, inicialize a matriz de forma crescente de 1 até 100:

```python
%%file lista9Exerc1.c

#include <stdio.h>

#define N 10

int main(){
    
    int matriz[N][N], valor = 1, i, j;

    // Inicializando a matriz - valor vai de 1 até 100

    for (i = 0; i < N; i++) {
        for (j = 0; j < N; j++) {
            matriz[i][j] = valor++;
        }
    }

    // Exibindo a matriz 

    for (i = 0; i < N; i++) {
        for (j = 0; j < N; j++) {
            printf("%3d ", matriz[i][j]);
        }
        printf("\n");
    }
    
    return 0;
}
```

2) Declare uma matriz de inteiros com 10 linhas e 10 colunas, inicialize a matriz de forma decrescente de 100 até 1

```python
%%file lista9Exer2.c

#include <stdio.h>

#define N 10

int main(){
    
    int matriz[N][N], valor = 100, i, j;

    // Inicializando a matriz - valor vai de 100 até 1

    for (i = 0; i < N; i++) {
        for (j = 0; j < N; j++) {
            matriz[i][j] = valor--;
        }
    }

    // Exibindo a matriz 

    for (i = 0; i < N; i++) {
        for (j = 0; j < N; j++) {
            printf("%3d ", matriz[i][j]);
        }
        printf("\n");
    }
    
    return 0;
}
```

3) Escreva um programa que irá receber do usuário uma matriz de números reais 3x3. Seu programa deve ser capaz de somar os elementos da diagonal principal dessa matriz e exibir o resultado ao usuário. Você também deverá multiplicar os elementos da diagonal secundária e exibir o resultado da multiplicação. Diagonal principal: i == j. Diagonal secundária: i + j == n - 1, onde n é o número de linhas da matriz.

```python
%%file lista9Exer3.c

#include <stdio.h>

#define N 3
#define M 3

int main(){

    float matriz[N][M];
    float soma = 0, mult = 1;
    int i, j;

    // Entrada de dados: recebendo os valores do usuário

    for (i = 0; i < N; i++) {
        for (j = 0; j < M; j++) {
            printf("informe o valor para matriz[%d][%d]\n", i, j);
            scanf("%f", &matriz[i][j]);
        }
    }

    // Operando na matriz

    for (i = 0; i < N; i++) {
        for (j = 0; j < M; j++) {            

            // Diagonal principal

            if (i == j) {
                soma += matriz[i][j];
            }

            // Diagonal secundária

            if (i + j == N - 1) {
                mult *= matriz[i][j];
            }
        }
    }

    // Exibindo o valor da soma e da multiplicação

    printf("soma: %.2f\n", soma);
    printf("multiplicação: %.2f\n", mult);

    return 0;
}
```

4) Implemente um algoritmo capaz de determinar se uma matriz quadrada qualquer é uma matriz triangular superior ou inferior:

```python
%%file lista9Exer4.c

#include <stdio.h>

#define N 3

int main(){
    
    int matriz[N][N], i, j, e_superior = 1, e_inferior = 1;
    
    // Inicializando a matriz - valores informador pelo usuário

    for (i = 0; i < N; i++) {
        for (j = 0; j < N; j++) {
            printf("informe o valor para matriz[%d][%d]: ", i, j);
            scanf("%d", &matriz[i][j]);
        }
    }

    // Exibindo a matriz informada em tela

    for (i = 0; i < N; i++) {
        for (j = 0; j < N; j++) {
            printf("%d ", matriz[i][j]);
        }
        printf("\n");
    }

    // Verificando se a matriz é triangular superior 

    for (i = 1; i < N; i++) {
        for (j = 0; j < i; j++) {
            if (matriz[i][j] != 0) {
                e_superior = 0;
            }
        }
    }

    // Verificando se a matriz é triangular superior 

    for (i = 0; i < N - 1; i++) {
        for (j = i + 1; j < N; j++) {
            if (matriz[i][j] != 0) {
                e_inferior = 0;
            }
        }
    }

    if (e_superior == 1) {
        printf("Matriz informada é uma matriz triangular SUPERIOR\n");
    }

    if (e_inferior == 1) {
        printf("Matriz informada é uma matriz triangular INFERIOR\n");
    }
    
    return 0;
}
```

5) Com base na matriz abaixo, conte o número de elementos ímpares que a mesma possui. Todo o elemento ímpar deve ser substituído pelo número 1 e todo o elemento par deve ser substituído pelo o número 0.

```python
%%file lista9Exer5.c

#include <stdio.h>

int main(){
    
    int matriz[4][4] = {1, 2, 4, 6, 8, 3, 2, 4, 6, 8, 5, 2, 4, 6, 8, 7}; 
    int i, j, cont = 0;
    
    printf("Matriz original:\n");

    for (i = 0; i < 4; i++) {
        for (j = 0; j < 4; j++) {
            printf("%d ", matriz[i][j]);
        }
        printf("\n");
    }

    // Verificando se a matriz é triangular superior 

    for (i = 0; i < 4; i++) {
        for (j = 0; j < 4; j++) {
            if (matriz[i][j] % 2 == 0) {    // testa se matriz[i][j] é par
                matriz[i][j] = 0;
            }
            else {
                matriz[i][j] = 1;
                cont++;
            }
        }
    }

    printf("Matriz modificada:\n");

    for (i = 0; i < 4; i++) {
        for (j = 0; j < 4; j++) {
            printf("%d ", matriz[i][j]);
        }
        printf("\n");
    }

    printf("Contagem de ímpares: %d\n", cont);
    
    return 0;
}
```

6) Escreva um programa para inicializar randomicamente uma matriz 5x5 com números inteiros. Na sequência realize as seguintes operações:
-   Troque a segunda linha e quinta linha;
-   Torque a primeira linha e a quarta coluna; 
-   Troque a diagonal principal e a secundária; 
-   Exiba a matriz ao final do processo de trocas.

```python
%%file lista9Exer6.c

#include <stdio.h>
#include <stdlib.h>
#include <time.h>

#define N 5

int main(){
    
    int matriz[N][N], i, j, k, aux;

    srand(time(NULL));
    
    // Inicializando a matriz com valore aleatórios (de 0 até 100)

    for (i = 0; i < N; i++) {
        for (j = 0; j < N; j++) {
            matriz[i][j] = rand() % 101;
        }
    }

    printf("Matriz original:\n");

    for (i = 0; i < N; i++) {
        for (j = 0; j < N; j++) {
            printf("%3d ", matriz[i][j]);
        }
        printf("\n");
    }

    // a) Troque a segunda linha (índice de linha 1) e quinta linha (índece de linha 4)
    
    for (j = 0; j < N; j++) {
        aux = matriz[1][j]; 
        matriz[1][j] = matriz[4][j];
        matriz[4][j] = aux;
    }

    printf("Matriz após operação A (troca da linha 1 pela linha 4):\n");

    for (i = 0; i < N; i++) {
        for (j = 0; j < N; j++) {
            printf("%3d ", matriz[i][j]);
        }
        printf("\n");
    }    

    // b) Torque a primeira linha (índice de linha 0) e a quarta coluna (índice de coluna 3)

    for (k = 0; k < N; k++) {
        aux = matriz[0][k]; 
        matriz[0][k] = matriz[k][3];
        matriz[k][3] = aux;
    }

    printf("Matriz após operação B (troca da linha 0 pela coluna 3):\n");

    for (i = 0; i < N; i++) {
        for (j = 0; j < N; j++) {
            printf("%3d ", matriz[i][j]);
        }
        printf("\n");
    }

    // c) Troque a diagonal principal e a secundária
 
    for (k = 0, j = N - 1; k < N; k++, j--) {
        aux = matriz[k][k]; 
        matriz[k][k] = matriz[k][j];
        matriz[k][j] = aux;
    }

    printf("Matriz após operação C (troca da diagonal principal pela secundária):\n");

    for (i = 0; i < N; i++) {
        for (j = 0; j < N; j++) {
            printf("%3d ", matriz[i][j]);
        }
        printf("\n");
    } 
    
    return 0;
}
```
